```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as py
import seaborn as san
data = pd.read_csv(r"C:/Users/Royal007Yash/Downloads/Python_Diwali_Sales_Analysis-main/Python_Diwali_Sales_Analysis-main/Diwali Sales Data.csv",encoding='ISO-8859-1')
data.shape
data.head(50)
data.info()
data.drop(['Status', 'unnamed1'], axis=1,inplace=True)
pd.isnull(data)
pd.isnull(data).sum()
data.dropna(inplace=True)
pd.isnull(data).sum()
data['Amount']=data['Amount'].astype('int')
data.info()
data.describe()
data[['Age','Amount']].describe()
```

```python
ax=san.countplot(x ='Gender' ,hue='Gender',data=data,palette='Set2')
py.title("Count of Male And Female")
for bars in ax.containers:
    ax.bar_label(bars)  
```

```python

sales_gender=data.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=True)   
san.barplot(x="Gender",y="Amount", data=sales_gender ,hue='Gender',palette='Set1')
py.title("Most Buyer according to Gender")
```

```python
ax = san.countplot(data=data, x="Age Group", hue='Gender', palette='Set2')
py.title("Age Group Distribution by Gender")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_group=data.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by="Amount",ascending=True)   
san.barplot(x="Age Group",y="Amount", data=sales_group ,hue='Age Group',palette='Set1')
py.title("Most Buyer according to Age Group")
```

```python
sales_state=data.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=True)
san.set(rc={'figure.figsize':(30,15)})
san.barplot(data=sales_state,x="State",y="Orders",hue="State",palette="deep")
py.title("StateWise Orders" ,fontsize=20)
```

```python
sales_state=data.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=True).head(5)
san.set(rc={'figure.figsize':(30,15)})
san.barplot(data=sales_state,x="State",y="Orders",hue="State",palette="deep")
py.title("Bottom 5 StateWise Orders" ,fontsize=20)
```
